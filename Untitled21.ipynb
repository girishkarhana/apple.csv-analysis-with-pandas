```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/apple.csv')
print(df)
```

```python
df.head()
```

```python
top_3_phones= df.nlargest(5, 'Star Rating')[['Product Name','Star Rating','Number Of Ratings']]
print(top_3_phones)
```

```python
top_3_phones= df.nsmallest(5, 'Star Rating')[['Product Name','Star Rating','Number Of Ratings']]
print(top_3_phones)
```

```python
highest_review=df.loc[df['Number Of Reviews'].idxmax(),['Product Name','Number Of Reviews']]
print(highest_review)
```

```python
highest_review=df.loc[df['Number Of Reviews'].idxmin(),['Product Name','Number Of Reviews']]
print(highest_review)
```

```python
x=df['Star Rating']
y=df['Number Of Reviews']
plt.subplot(2,1,1)
plt.plot(x, y, label='Sale Price vs Number Of Reviews', color='blue')
plt.xlabel('Sale Price')
plt.ylabel('Number Of Reviews')
plt.title('Sale Price vs Number Of Reviews')
plt.legend()

plt.show()

plt.subplot(2,1,2)
plt.plot(df['Star Rating'], df['Sale Price'], label='Sale Price vs Number Of Ratings', color='blue')
plt.ylabel('Sale Price')
plt.xlabel('Star Rating')
plt.title('Sale Price vs Number Of Reviews')
plt.legend()

plt.show()
```

```python
plt.subplot(2,1,1)
x=df['Number Of Ratings']
y1=df['Discount Percentage']
plt.plot(x,y1,linewidth=2,label="y3",color="blue")
plt.title("Number Of Ratings vs Discount Percentage")
plt.legend()
plt.show()
plt.subplot(2,1,2)
x=df['Number Of Reviews']
y1=df['Discount Percentage']
plt.plot(x,y1,linewidth=2,label="y3",color="blue")
plt.title("Number Of Reviews vs Discount Percentage")
plt.legend()
plt.show()
```

```python
x = df['Product Name'].head()
y1 = df['Number Of Ratings'].head()
y = df['Sale Price'].head()
plt.figure(figsize=(10, 6))
plt.bar(x, y1, color="blue", alpha=0.7, label="Number Of Ratings")
plt.bar(x, y, color="red", alpha=0.5, label="Sale Price")

plt.title("Number of Ratings vs Sale Price", fontsize=14)
plt.xlabel("Product Name", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
